```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_highcharts.display import display_charts

# Get needed data file
input_file = '..\data\india-districts-census-2011.csv'
df = pd.read_csv(input_file)
```

```python
df.describe()
```

```python
# Task 05 - State wise male and female population in India
df_population = df[['State Name', 'Male','Female', 'Population']]
mf_population =  df_population.groupby(['State Name'])[['Male','Female','Population']].sum().rename(columns={'Population':'TOTAL'})
display_charts(mf_population, title="India's population - Male and Females", kind="bar", figsize = (1000, 700))
```

```python
# 07 - Population of different types of age groups in India

df_age_group = df.groupby(['State Name'])[['Age_Group_0_29','Age_Group_30_49','Age_Group_50']].sum().rename(columns={'Age_Group_0_29':'0-29 YEARS','Age_Group_30_49':'30-49 YEARS','Age_Group_50':'50 YEARS & ABOVE'})
display_charts(df_age_group, title="State wise different age groups in India", kind="barh",figsize = (1000, 700))
```

```python
# Task 09 - Religion's distribution in India

df_reg = df[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists','Jains','Others_Religions','Religion_Not_Stated']].sum()
df_religion = pd.DataFrame(df_reg)

df_religion.rename(columns={0:'RELIGIONS'})

display_charts(df_religion, kind='pie', title="Religions in India") 
```
